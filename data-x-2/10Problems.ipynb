```python
x1 = np.array([2,3,4,5,2.5,3.5,4.5,5.5])
y = np.array([1.1,1.5,1.9,2.5,1.25,1.75,2.25,2.75])
```

Predictor using first 4 values

```python
x1_train = np.array(x1[:4]).reshape(-1,1)
y_train = np.array(y[:4]).reshape(-1,1)
```

```python
lr = 0.01
iterations = 10
theta = np.random.randn(2,1)
print(theta)
```

```python
def gradient_descent(X,y,theta,alpha = 0.01, itern = 100):
    m = len(y)
    cost_hist = np.zeros(itern)
    theta_hist = np.zeros((itern,2))
    for i in range(itern):
        prediction = np.dot(X, theta)
        theta = theta - (1/m)*alpha*(X.T.dot(prediction-y))
        theta_hist[i,:] = theta.T
        cost_hist[i] = cal_cost(theta,X,y)
    return theta, cost_hist, theta_hist
```

i) Write out the mean squared error loss function

```python
def cal_cost(theta,X,y):
    m = len(y)
    predictions = np.dot(X, theta)
    cost = np.average(np.square(predictions - y))/2
    return cost
```

```python
X_b = np.c_[np.ones((len(x1_train),1)),x1_train]
theta, cost_history, theta_history = gradient_descent(X_b,y_train,theta,lr,iterations)
```

```python
mse_vals = np.zeros(shape=(w0.size, w1.size))
theta_r = np.random.randn(2,1)
for i,value1 in enumerate(w0):
    for j,value2 in enumerate(w1):
        theta_r[0] = value1
        theta_r[1] = value2
        mse_vals[i, j] = cal_cost(theta_r,X_b,y_train)
```

```python
plt.contour(X1, X2, mse_vals)
plt.title("Contour Plot of Gradient Descent")
plt.xlabel("w0")
plt.ylabel("w1")
plt.show()
```
